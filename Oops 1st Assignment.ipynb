```python
class bank_balance:
    def __init__(self,name,balance,account_no):
        self.name = name
        self.balance = balance
        self.account_no = account_no


    def display_balance(self):
        print(f"Your balance for account no. {self.account_no} is",self.balance)

    def debit(self,amount):
        self.balance+=amount
        print(f"Deposit of Rs. {amount} is successful")
        self.display_balance()

    def credit(self,amount):
        self.balance-=amount
        print(f"Withdraw of Rs. {amount} is successful." )
        self.display_balance()

client = bank_balance("Dipanjan", 500, 12345)
client.display_balance()
client.debit(500)
client.credit(300)
```

```python
class employee:
    def __init__(self,employee_id,name,salary,bonus):
        self.employee_id = employee_id
        self.name = name
        self.salary = salary
        self.bonus = bonus

    
    def yearly_bonus(self,yearly_present):
        
        if yearly_present in range (200,300):
            self.bonus+= 5000
        elif (yearly_present in range(100,199)):
            self.bonus+= 1000
        else:
            print("Not in bonus criteria")
        self.show_bonus()
    
    def show_bonus(self):
        print(f"Dear {self.name} your id {self.employee_id} and your salary is {self.salary} bonus is {self.bonus} and total payment is {self.salary+self.bonus}")

employee1 = employee(12345,"Dipanjan Pathak",60000,0)
employee1.yearly_bonus(250)
```

```python
class VehicleRental:
    def __init__(self):
        self.available_vehicles = {}

    def rent_vehicle(self, vehicle_id, vehicle_details):
        if vehicle_id not in self.available_vehicles:
            self.available_vehicles[vehicle_id] = vehicle_details
            print(f"Vehicle {vehicle_id} rented successfully.")
        else:
            print(f"Vehicle {vehicle_id} is already rented.")

    def return_vehicle(self, vehicle_id):
        if vehicle_id in self.available_vehicles:
            del self.available_vehicles[vehicle_id]
            print(f"Vehicle {vehicle_id} returned successfully.")
        else:
            print(f"Vehicle {vehicle_id} was not rented.")

    def display_available_vehicles(self):
        if self.available_vehicles:
            print("Available Vehicles:")
            for vehicle_id, details in self.available_vehicles.items():
                print(f"ID: {vehicle_id}, Details: {details}")
        else:
            print("No vehicles available for rent.")
```

```python

class library:
    def __init__(self):
        self.booklists = {}
        

    def add_books(self,book_id):
        
        if book_id not in self.booklists:
            self.booklists[book_id]= "in_stock"
        print(f"{book_id} added to the library!")
        

    def borrow_books(self):
        book_borrow = input("Enter book id to be borrowed: ")
        if book_borrow in self.booklists and self.booklists[book_borrow] == "in_stock":
            print(f"{book_borrow} Book borrowed!")
            self.booklists[book_borrow] = "borrowed"
        else:
            print("Book not available for borrowing.")

    def return_book(self):
        return_book = input("Enter book id to be returned: ")
        if return_book in self.booklists and self.booklists[return_book] == "borrowed":
            print(f"{return_book} Book returned!")
            self.booklists[return_book] = "in_stock"


    
    def available_books(self):
        stock_book_list = []
        for book_id,status in  self.booklists.items():
            if status == "in_stock":
             stock_book_list.append(book_id)
        print(f"Available book are {stock_book_list}")


library_initialize = library()
while True:
 user_choice = int(input("Enter your choice:\n1. for book add\n2. borrow books\n3. return book\n4. display book\n5. for quite\n"))
 if user_choice == 5:
    break
 elif user_choice ==1:
    user_input = input("Enter the book id: ")
    library_initialize.add_books(user_input)
 elif user_choice == 2:
    library_initialize.borrow_books()
 elif user_choice == 3:
    library_initialize.return_book()
 elif user_choice ==4:
    library_initialize.available_books()
```

```python
class inventory:
    def __init__(self):
        self.product = {}
            
    def display_product(self):
        if len(self.product)==0:
            print("Opps! Inventory is empty, Please add product: ")
        else:
           for key, value in self.product.items():
             print(f"product {key} is in stock of {value}")


    def add_product(self):
        product_name_input = input("Enter product name: ")
        product_quantity_input = input("Enter product quantity") 
        self.product[product_name_input] = product_quantity_input
        self.display_product()
        

user = inventory()
while True:
    user_input = input ("Enter you choice:\n1. Add product or update product quantity\n2. Display Stock\n3. For quite\n")
    if user_input == '3':
        break
    elif user_input =='1':
        user.add_product()
    elif user_input=='2':
        user.display_product()
    else:
        print("invalid Input")
```

```python
class shape:
    def __init__(self,length = None,width= None,height= None):
        self.length = length
        self.width = width
        self.height = height
   
    def rectangle_area(self):
        if self.length !=None and self.width != None:
            return int(self.length)*int(self.width)
        elif self.length is not None:
            return int(self.length)**2
        elif self.width is not None:
            return int(self.width)**2
        
    def triangular_area(self):
        if self.length != None and self.width != None and self.height != None:
            s = (self.length+self.width+self.height)/2
            return (s*(s-self.length)*(s-self.width)*(s-self.height))**0.5
        else:
            print("Enter all required input")
        
    
a = shape()
while True:
     user_input = input("Enter your choice: \n1. Area of rectangle \n2. Area of triangle \n3. Quite\n")
     if user_input == '3':
         break
     elif user_input =='1':
        length = input("Enter the length: ")
        width = input("Enter the width: ")
        a.length = length
        a.width = width
        rec_area = a.rectangle_area()
        print("Area is",rec_area)
        
     elif user_input =='2':
        length = int(input("Enter the 1st side: "))
        width = int(input("Enter the 2nd side: "))
        height = int(input("Enter the 3rd side: "))
        a.length = length
        a.width = width
        a.height = height
        tri_area = a.triangular_area()
        print("Area is",tri_area)
     else:
         print("Opps! invalid input")
```

```python
class StudentManagement:
    def __init__(self):
        self.grades = {}

    def average_grade(self):
        while True:
            student_id = input("Enter student ID or press 'Q' to quit: ")
            if student_id.lower() == 'q':
                break
            else:
                student_name = input("Enter student Name: ")
                total_marks = 0
                count = 0
                while True:
                    subject_name = input("Enter subject name or press 'Q' to quit: ")
                    if subject_name.lower() == 'q':
                        break
                    else:
                        subject_grade = int(input("Enter subject grade: "))
                        if student_id not in self.grades:
                            self.grades[student_id] = {'name': student_name, 'subjects': {}}
                        self.grades[student_id]['subjects'][subject_name] = subject_grade
                        total_marks += subject_grade
                        count += 1
                if count > 0:
                    average = total_marks / count
                    self.display_details(student_id, student_name, average)

    def display_details(self, student_id, student_name, average_grade):
        print(f"Student name: {student_name}, ID: {student_id}, and average grade is {average_grade}")

student_manager = StudentManagement()
student_manager.average_grade()
```

```python
class SocialMediaProfile:
    def __init__(self, username):
        self.username = username
        self.posts = []

    def add_post(self, post_content):
        self.posts.append(post_content)

    def display_posts(self):
        if self.posts:
            print(f"Posts by {self.username}:")
            for post in self.posts:
                print(post)
        else:
            print(f"No posts found for {self.username}")

    def search_posts(self, keyword):
        matching_posts = [post for post in self.posts if keyword in post]
        if matching_posts:
            print(f"Posts containing '{keyword}' by {self.username}:")
            for post in matching_posts:
                print(post)
        else:
            print(f"No posts containing '{keyword}' found for {self.username}")


profile = SocialMediaProfile("example_user")
profile.add_post("Hello world! This is my first post.")
profile.add_post("Just had a great day outdoors.")
profile.add_post("Enjoying some quality time with friends.")
profile.display_posts()

print("\nSearching for posts containing 'great':")
profile.search_posts("great")
```

```python
# 1. making class
class ToDoList:
    def __init__(self):
        self.task = {"date1":{'pending':['Bath','Eat'],'completed':['Brush','walk']}}

# 2. add task function
    def add_task(self):
        while True:
         task_date = input ("Enter date or enter 'Q' for quite: ")
         if task_date.lower()=='q':
            break
         else:
          if task_date not in self.task:
           self.task[task_date] = {'pending': [], 'completed': []}
         while True:
            task_input = input("Enter task to add or 'Q' for Quite: ")
            if task_input.lower() == 'q':
                break
            else:
                self.task[task_date]['pending'].append(task_input)
        

# 3. mark task as completed function
    def mark_task(self):
        task_date = input ("Enter date: ")
        while True:
            task_input = input("Enter task to complete or 'Q' for Quite: ")
            if task_input.lower() == 'q':
                break
            elif task_input in self.task[task_date]['pending']:
                self.task[task_date]['completed'].append(task_input)
                self.task[task_date]['pending'].remove(task_input)
            else:
                print("Invalid Input")
        


# 4. display pending task function
    def display_pending(self):
        display_date = input("Enter the Date: ")
        for i in self.task[display_date]['pending']:
         print(i)


a = ToDoList()
while True:
    user_input = input("Enter your choice:\n1. Add task\n2.Mark task as completed\n3. Display pending task\n4. 'Q' for quite\n")
    if user_input.lower() =='q':
        break
    elif user_input =='1':
        a.add_task()
    elif user_input =='2':
        a.mark_task()
    elif user_input =='3':
        a.display_pending()
    else:
        print("invalid input!")
```
